```python
import pandas as pd
from kafka import KafkaProducer
import json
import time
import os

# Your exact cached path from Lab 1
dataset_path = os.path.expanduser(
    "~/.cache/kagglehub/datasets/grouplens/movielens-latest-small/versions/2"
)

# Initialize Kafka Producer (assuming default localhost port)
producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8"),
)

print("1. Loading datasets...")
movies_df = pd.read_csv(f"{dataset_path}/movies.csv")
# Sort streams by timestamp to simulate realistic event progression
ratings_df = pd.read_csv(f"{dataset_path}/ratings.csv").sort_values("timestamp")
tags_df = pd.read_csv(f"{dataset_path}/tags.csv").sort_values("timestamp")

print("2. Pushing static Movies data to Kafka...")
for _, row in movies_df.iterrows():
    producer.send("Lab1_movies", row.to_dict())
producer.flush()
print("Movies pushed successfully!")

print("2. Pushing static Ratings data to Kafka...")
for _, row in ratings_df.iterrows():
    producer.send("Lab1_ratings", row.to_dict())
producer.flush()
print("Ratings pushed successfully!")

print("2. Pushing static Tags data to Kafka...")
for _, row in tags_df.iterrows():
    producer.send("Lab1_tags", row.to_dict())
producer.flush()
print("Tags pushed successfully!")
```
